```python
import pandas as pd
import os

def get_filelist(dir):
    for home, dirs, files in os.walk(dir):
        files_list=[]
        for filename in files:
#             print(filename)
            if filename.split('.')[-1]=='csv':
                if filename.split('.')[-2][-2:]=='fk':
                    files_list.append(filename)
    return files_list
```

```python
csvdir='F:\\2017\\2017Csv'
fk_files=get_filelist(csvdir)
```

```python
data=pd.read_csv(os.path.join(csvdir,fk_files[0]))
for file in fk_files[1:]:
    data=data.append(pd.read_csv(os.path.join(csvdir,file)))
                     
data.to_csv('all_city.csv',index=None,encoding='utf_8_sig')
data.head()
```

```python
#进行删除操作
def get_con(a,b):
    if a==b:
        return 0
    else:
        return 1

data['Cond'] = data.apply(lambda x: get_con(x['TARGET'],x['SOURCE']), axis=1)
```

```python
data.shape
```

```python
data[data['Cond']==1].to_csv('cond_city.csv',index=None,encoding='utf_8_sig')
```
